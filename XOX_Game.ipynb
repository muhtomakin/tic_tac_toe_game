```python
import random

def display_board(board):
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   %c   |   %c   |   %c   |" % (board[1],board[2],board[3]))
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   %c   |   %c   |   %c   |" % (board[4],board[5],board[6]))
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print("|   %c   |   %c   |   %c   |" % (board[7],board[8],board[9]))
    print("|       |       |       |")
    print("+-------+-------+-------+")
    #return
```

```python
def enter_move(board):
    display_board(board)
    turn = 1
    while True:
        #move = input("Enter your move: ")
        #m=int(move)
        while True:
            try:
                move = input("Enter your move: ")
                m=int(move)
                break
            except ValueError:
                print("Please enter number!!!")
                #continue
            #print(sign)
        if turn %2 == 0:
            for i in board:
                if move == i:
                    board[m] = "X"
        elif turn %2 == 1:
            for i in board:
                if move == i:
                    board[m] = "O"
        display_board(board)
        turn += 1
        if victory_for(board):
            if turn %2 == 1:
                print("Player {} won.".format("X"))
                break
            elif turn %2 == 0:
                print("Player {} won.".format("O"))
                break
        if draw(board):
            print("Draw.")
            break
```

```python
def victory_for(board):
    if(board[1]==board[2] and board[1]==board[3] and board[2]==board[3]):
            return("Player {} won.".format(board[1]))
    elif(board[4]==board[5] and board[4]==board[6] and board[5]==board[6]):
            return("Player {} won.".format(board[4]))
    elif(board[7]==board[8] and board[7]==board[9] and board[8]==board[9]):
            return("Player {} won.".format(board[7]))
    elif(board[1]==board[5] and board[1]==board[9] and board[5]==board[9]):
            return("Player {} won.".format(board[1]))
    elif(board[1]==board[4] and board[1]==board[7] and board[4]==board[7]):
            return("Player {} won.".format(board[1]))
    elif(board[2]==board[5] and board[2]==board[8] and board[5]==board[8]):
            return("Player {} won.".format(board[2]))
    elif(board[3]==board[5] and board[3]==board[7] and board[5]==board[7]):
            return("Player {} won.".format(board[3]))
    elif(board[3]==board[6] and board[3]==board[9] and board[6]==board[9]):
            return("Player {} won.".format(board[3]))
```

```python
def draw(board):
    if (board[1] == 'X' or board[1] == 'O') and (board[2] == 'X' or board[2] == 'O') \
        and (board[3] == 'X' or board[3] == 'O') and (board[4] == 'X' or board[4] == 'O') \
        and (board[5] == 'X' or board[5] == 'O') and (board[6] == 'X' or board[6] == 'O') \
        and (board[7] == 'X' or board[7] == 'O') and (board[8] == 'X' or board[8] == 'O') \
        and (board[9] == 'X' or board[9] == 'O'):
        return ("Draw")
```
